```python
# a map function is a special type of funtions that applies to the every item in the list and then returns a map object that is iterable 

def square(x):
    return x*x

number = [1,2,3,4,5,6,7,8]

squared_numbers =list(map(square,number))
print(squared_numbers)
```

```python
# lambda function with the map function

square_num  = list(map(lambda x:x*x,number))
print(square_num)
```

```python
# mapping mutiple iterables

num1 =[1,2,3]
num2 =[4,5,6]

add_number = list(map(lambda x,y:x+y,num1,num2))
print(add_number)
```

```python
# map to convert a list of string to numbers
str_num = ["1","2","3","4"]
numbers_converted =list( map(int,str_num))
print(numbers_converted)
```

```python
# filter function is used to filter out items from a list based on a condition

def even(num):
    if num%2==0:
        return True

lst = [1,2,3,4,5,6,6,7,8,9,0,10,11,1,33,4,44]

even_list=list(filter(even,lst))
print(even_list)
```

```python
# filter in dictionary

people =[
    {'name':'sahil','age':22},
    {'name':"akash",'age':20},
]

def above20(people):
    return people['age']>20

list_people_above_20 = list(filter(above20,people))
print(list_people_above_20)
```
