```python
# sets are built in data types in python that are used to store the unique elements. They are not ordered and does not folloe any specific order.
my_set= {}
print(type(my_set))
print(my_set)
```

```python
# basic set operations

#adding 
my_set.add(10)
my_set
my_set.add(10)
my_set

#removing the elements
my_set.add(3)
my_set.remove(3)
my_set

# let's say that the element that we want to remove dose't exist use discard method it doesn't give an error
my_set.discard(10)
my_set

# removing a random element
popped_ele = my_set.pop()
print(popped_ele)


```

```python
# clear all the elements
my_set.clear()
my_set
```

```python
# set membership set
my_set = (1,2,3,4,5,6)
print(3 in my_set)
```

```python
# maths operations
set1 = {1,2,3,4}
set2 = {4,5,6,7}

##union operation
union_set  = set1.union(set2)
union_set

#interseciton operation
common_ele = set1.intersection(set2)
common_ele

#symmetric difference 
symmetric_diff = set1.symmetric_difference(set2)
symmetric_diff
```
