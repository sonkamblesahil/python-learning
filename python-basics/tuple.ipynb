```python
## tuples are immmutable and cannot be assigned with new values

tup = (1,2,3,"sahil",True)
print(tup)
```

```python
print(tup[-1])
lst = list(tup)
print(lst)
```

```python
rev = tup[::-1]
print(rev)
```

```python
num = (1,2,3,4)
concat = num+tup
print(concat)
```

```python
num*3
```

```python
num[0]= 10  #this is called as immutability of tuple as tuple object dosen't suppport assignment
print(num)
```

```python
#tuple methods
print(num.count(2))
print(num.index(3))
```

```python
#packing and unpacking a tuple

#packing
h = 1,"sahil","this",True
print(h)

#unpacking
a,b,c,d = h
print(a,b,c,d)

#unpacking with *
numbers = (1,2,3,4,5,6)
one,*second,last = numbers
print(one,second,last)

```

```python
# we can add a tuple inside a list
new = [[1,2,3,4],("sahil","sonkamble")]
print(new[1][:])
```
