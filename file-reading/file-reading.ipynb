```python
# reading the whole file

with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# reading file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # .strip() removes the newline character
```

```python
# writing a file the entire content is overridden

with open('example.txt','w') as file:
    file.write("Hello added! \n")
    file.write("this sia new line \n")
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# write a file without overriding

with open('example.txt','a') as file:
    file.write("\nnew line got appended with previous data \n")
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())

```

```python
# writign a list of lines on a file
list = ['\nhey',"\nhello","\nhi"]
with open('example.txt','a') as file:
    file.writelines(list)
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```
