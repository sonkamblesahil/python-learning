```python
# a function is a block of code that performs a particular task it increases the reusability and helps in organizing the code

def print_hello():
    print("Hello")

print_hello()
```

```python
def even_odd(num):
    if num%2==0:
        print("number is even")
    else:
        print("number is odd")

even_odd(2)
even_odd(11)
```

```python
# functions with multiple parametes

def add(num1,num2):
    return num1+num2

sum  = add(8,10)
print(sum)
```

```python
# default parameters

def multiply(number =2):
    print(number*8)
multiply()
multiply(4)
multiply(10)
```

```python
# variable length arguments
# positional and keywords arguments

def print_numbers(*args): # args is the standard we can give any name
    for num in args:
        print(num)

print_numbers(1,2,3,34,4,4,5,0)

```

```python
# keywords arguments

def print_details(**kwargs): # kwargs is a standard we can give any name
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="sahil",age=32)
```

```python
def print_detailsindetails(*args,**kwargs):
    print("positional arguments")
    for val in args:
        print(f"{val}")
    print("keyword argumnets")
    for key,val in kwargs.items():
        print(f"{key}:{val}")

print_detailsindetails(1,2,3,"sahil",name="sahil",age=32)
```
