```python
# Dictionaries store data as key-value pairs.
# Keys must be hashable; values can be any type.
# Dictionaries preserve insertion order (Python 3.7+).

my = {}
print(type(my))      # <class 'dict'>

my_set = set()
print(type(my_set))  # <class 'set'>
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student ={
    "name":"sahil",
    "age":22,
    "grade":"A+"
}
print(student)
```

```python
## accessing the dictionary elements
student["grade"]="B+" 
print(student["grade"])

print(student.get("name"))
print(student.get("last_name")) # returns none if that key is not present
print(student.get("last_name","Not available")) # instead of None prints the default values of not available

```

```python
# modifying dictionary elements -> dictionaries are mutable 

 # updating the existing key
student["age"]=23 
print(student)

# adding  new key
student["last_name"]="sonkamble" 
print(student)

# deleting the entire key-value
del student["name"]
print(student)


```

```python
#common dictionary methods

# get all the keys
keys = student.keys()
print(keys)

# get all the values
values =  student.values()
print(values)

# get all key value pairs
items = list(student.items())
print(items)
print(items[0][0]) # prints age
```

```python
# copying dictionary refers to same object in memory
student_cpy  = student
print(student)
print(student_cpy)
```

```python
student["name"]="sahil1"
print(student)
print(student_cpy)
```

```python
# shallow copy
 
student_scpy = student.copy() # allocate a different memory for that object 
print(student_scpy)
```

```python
student["name"] = "vishal"
print(student_scpy) # name dosent change as we have different memory location
```

```python
# iterating over dictionaries, values, keys and items

# using loops

# iterating over keys
for keys in student.keys():
    print(keys)

# iteraton over values
for val in student.values():
    print(val)

# iterating over items

for key,val in student.items():
    print(f"{key}:{val}")
```

```python
# nested dictionaries

students ={

    "student1":{
        "name":"sahil",
        "age":22,
    },
    "student2" : {
        "name":"akash",
        "age":19
    },
}

print(students)
print(students["student1"])
print(students["student1"]["name"])
```

```python
# looping over teh nested list

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,val in student_info.items():
        print(f"{key}:{val}")
```

```python
# dictionary comprehension

squares = { x:x*x for x in range(10)}
print(squares)
```

```python
# conditional dictionary comprehension

even = { x:x for x in range(10) if x%2==0}
print(even)
```

```python
# merginf of 2 dictionaries
dict1 ={"a":1,"b":2}
dict2 ={"b":3,"d":5}
merged  = {**dict1,**dict2}
print(merged)
```
